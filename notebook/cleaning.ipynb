# Filling in missing values

```python
data.isnull().sum()
```

```python
data['Arrival Delay'].fillna(data['Arrival Delay'].median(axis = 0), inplace = True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

# Handling categorical features

```python
data.Satisfaction.value_counts()
```

```python
def find(x):
    if x == 'Satisfied':
        return 0 
    else:
        return 1 
data['Satisfaction'] = data.Satisfaction.apply(lambda x: find(x))
data.Satisfaction.value_counts()
```

```python
sns.set_theme()
```

```python
data.select_dtypes(include = ['object'])
```

```python
data.Class.value_counts()
```

```python
def finder(x):
    if x == 'Business':
        return 1 
    elif x == 'Economy':
        return 2 
    else:
        return 3     
data['Class'] = data.Class.apply(lambda x: finder(x))
data.Class.value_counts()
```

```python
def finder(x):
    if x == 'Male':
        return 1 
    elif x == 'Female':
        return 2 
    else:
        return 0 
data['Gender'] = data.Gender.apply(lambda x: finder(x))
data.Gender.value_counts()
```
